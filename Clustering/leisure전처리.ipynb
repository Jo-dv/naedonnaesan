```python
import pandas as pd
```

```python
df1 = pd.read_csv('편의시설 변경 버전.csv')
df1.head(3)
```

```python
df = pd.read_csv('dataset/leisure/서울시 체력단련장업 인허가 정보.csv',encoding='cp949')
df.head(3)
```

```python
df.columns
```

```python
df = df[['도로명주소','상세영업상태명']]
```

```python
df
```

```python
df['법정동'] = df['도로명주소'].str.extract('.*\(([^)]+)\)$')
df = df[['상세영업상태명','법정동']]
```

```python
df
```

```python
# '법정동' 컬럼에서 '동'으로 끝나는 부분만 추출
df['법정동'] = df['법정동'].str.extract('([가-힣]+동)')
```

```python
df
```

```python
df = df[(df['상세영업상태명'] == '영업중') & pd.notnull(df['법정동'])]
```

```python
df
```

```python
df = df['법정동'].value_counts()
```

```python
df
```

```python
merged_df = pd.merge(df1, df, left_on='법정동명', right_on='법정동', how='left')
```

```python
merged_df
```

```python
merged_df = merged_df.rename(columns={'count': '헬스시설'})
merged_df['헬스시설'].fillna(0, inplace=True)
merged_df
```

```python
merged_df
```

```python
merged_df['법정동명'] == ''
```

```python
merged_df.to_csv('편의시설 변경 버전.csv', index=False)
```
