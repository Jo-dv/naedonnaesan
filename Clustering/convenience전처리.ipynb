```python
import pandas as pd
```

```python
df1 = pd.read_csv('편의시설 변경 버전.csv')
df1.head(3)
```

```python
df = pd.read_csv('dataset/convenience/서울시 대규모점포 인허가 정보.csv', encoding='cp949', encoding_errors='ignore')
df
```

```python
df.columns
```

```python
df = df[['도로명주소','상세영업상태명','사업장명', '업태구분명']]
```

```python
df
```

```python
df = df[df['상세영업상태명'] == '정상영업']
```

```python
df['업태구분명'].unique()
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df
```

```python
df = df[df['업태구분명'] != '전문점']
```

```python
df['법정동'] = df['도로명주소'].str.extract('.*\(([^)]+)\)$')
```

```python
df = df[pd.notnull(df['법정동'])]
```

```python
df['법정동']
```

```python
df
```

```python
df = df.drop(axis=1, columns='도로명주소')
```

```python
df = df.drop(axis=1, columns=['업태구분명','상세영업상태명'])
```

```python
df
```

```python
df['법정동'] = df['법정동'].str.split(',').str[0]
```

```python
df
```

```python
grouped_df = df.groupby(['법정동']).size().reset_index(name='마트')
# grouped_df.sort_values(ascending=False, by='개수')
grouped_df
```

```python
merged_df = pd.merge(df1, grouped_df, left_on='법정동명', right_on='법정동', how='left')
```

```python
merged_df
```

```python
merged_df = merged_df.drop(axis=1, columns='법정동')
merged_df
```

```python
merged_df = merged_df.fillna(0)
```

```python
merged_df.columns
```

```python
merged_df.to_csv('편의시설 변경 버전.csv', index=False)
```
