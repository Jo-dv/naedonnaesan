```python
import pandas as pd
```

```python
df = pd.read_csv("편의시설 변경 버전.csv", encoding='cp949')
```

```python
df['여가'] = df['헬스시설2'] +  df['영화상영관']
```

```python
df = df[['법정동명','헬스시설2','영화상영관','여가']]
```

## 인덱스 1씩 미루기

```python
df['dong_id'] = df.index + 1
df
```

```python
df = df.drop(axis=1, columns=['법정동명'])
```

```python
df['헬스시설2'] = df['헬스시설2'].astype(int)
df['영화상영관'] = df['영화상영관'].astype(int)
df['여가'] = df['여가'].astype(int)
```
