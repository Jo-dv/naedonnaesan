qcut으로 각 컬럼별로 데이터를 5개의 클래스로 나누기

```python
for col in df1.columns:
    df1[col] = df1[col].astype(int)
    
```

```python
df2 = df1
```

```python
df2
```

```python
df2['치안'] = pd.qcut(df1['치안'], 2, labels=[1, 2])
df2['여가'] = pd.qcut(df1['여가'], 2, labels=[1, 2])
```

```python
# df2['치안'] = pd.qcut(df1['치안'], 2, labels=[1, 2])
# df2['여가'] = pd.qcut(df1['여가'], 2, labels=[1, 2])
df2['일반음식점'] = pd.qcut(df1['일반음식점'], 5, labels=[1, 2, 3, 4, 5])
df2['보건'] = pd.qcut(df1['보건'], 5, labels=[1, 2, 3, 4, 5])
df2['편의시설'] = pd.qcut(df1['편의시설'], 5, labels=[1, 2, 3, 4, 5])
# df2['지하철역'] = pd.qcut(df1['지하철역'], 2, labels=[1, 2])
# df2['버스정류소'] = pd.qcut(df1['버스정류소'], 5, labels=[1, 2, 3, 4, 5])
df2['대중교통'] = pd.qcut(df1['대중교통'], 5, labels=[1, 2, 3, 4, 5])
df2['카페'] = pd.qcut(df1['카페'], 5, labels=[1, 2, 3, 4, 5])
```

```python
df2
```

MinMaxScaler는 이상하게나옴.
StandardScaler 해보자.

```python
df_standard = StandardScaler().fit_transform(df2)
df_standard = pd.DataFrame(df_standard)
df_standard.describe().round(2)
```

```python
# PCA 주성분분석
pca_test = PCA(random_state=1107)
X_p = pca_test.fit_transform(df_standard)


percent_variance = np.round(pca_test.explained_variance_ratio_* 100, decimals =2)
columns = []
for i in range(len(percent_variance)):
    columns.append(f'PC{i+1}')
    

ax = plt.bar(x = range(len(percent_variance)), height=percent_variance, tick_label=columns)
plt.ylabel('Percentate of Variance Explained')
plt.xlabel('Principal Component')
plt.title('PCA Scree Plot')
plt.show()
pd.Series(np.cumsum(pca_test.explained_variance_ratio_))
```

```python
pca = PCA(n_components=4)
pc = pca.fit_transform(df_standard)
pcd = pd.DataFrame(pc)
pcd
```

```python
from sklearn.cluster import KMeans
```

```python
# 엘보우 방법을 사용하여 최적의 K값 탐색
sse = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=5)
    kmeans.fit(df_standard)
    sse.append(kmeans.inertia_)

# SSE 그래프 그리기
plt.plot(range(1, 11), sse, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('SSE')
plt.show()
```

```python
km = KMeans(n_clusters=4, random_state=5)
km.fit_transform(df_standard)
df2['labels'] = km.labels_
```

```python
df2['labels'].value_counts()
```

```python
## 색깔추가
palette1 = sns.color_palette('deep',10)
```

```python
plt.figure(figsize=(12, 7))
sns.countplot(x=km.labels_, palette=palette1)
plt.title('K-Means Clustering', fontsize=15)
plt.show()
```

```python
from sklearn.metrics import silhouette_samples, silhouette_score
score = silhouette_score(pcd, km.labels_)
score
```
