# GeoCoding - 주소 to 좌표

```python
import requests, json
from tqdm.notebook import tqdm
from requests.adapters import HTTPAdapter
from urllib3.util.retry import Retry
```

```python
result = []
```

```python
def get_cordinates(addresses):
    session = requests.Session()
    retry = Retry(connect=3, backoff_factor=0.5)
    adapter = HTTPAdapter(max_retries=retry)
    session.mount('https://', adapter)
    headers = {"Authorization": "KakaoAK ba25a018cf0921215c5f8c824762526e"}
    params = {"query": ""}
    
    for i, address in tqdm(enumerate(addresses)):
        url = "https://dapi.kakao.com/v2/local/search/address.json"
        params["query"] = address
        api_json = session.get(url, headers=headers, params=params)
        result.append((i, json.loads(api_json.text)))

    return result
```

```python
addresses = df["지번주소"].tolist()
addresses
```

```python
get_cordinates(addresses)
```
