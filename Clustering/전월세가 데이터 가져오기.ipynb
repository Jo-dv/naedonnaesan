```python
import pandas as pd
```

```python
df = pd.read_csv("dataset/서울특별시_전월세가_2023.csv", )
df1 = pd.read_csv("dataset/서울시 부동산 전월세가 정보_2022.csv", encoding="cp949")
# df2 = pd.read_csv("dataset/서울시 부동산 전월세가 정보_2021.csv", encoding="cp949")
# df = pd.read_csv("dataset/서울특별시_전월세가_2020.csv")
# df = pd.read_csv("dataset/서울특별시_전월세가_2019.csv")
```

```python
df1 = df1[['접수연도', '자치구명','법정동명','본번','부번','층','전월세 구분','임대면적(㎡)','보증금(만원)','임대료(만원)','건물명','건축년도','건물용도']]
# df2 = df2[['접수연도', '자치구명','법정동명','본번','부번','층','전월세 구분','임대면적(㎡)','보증금(만원)','임대료(만원)','건물명','건축년도','건물용도']]
df = df[['접수년도', '자치구명','법정동명','본번','부번','층','전월세구분','임대면적','보증금(만원)','임대료(만원)','건물명','건축년도','건물용도']]
```

```python
df1 = df1.rename(columns={'전월세 구분': '전월세구분'})
# df2 = df2.rename(columns={'전월세 구분': '전월세구분'})

df = df.rename(columns={'임대면적': '임대면적(㎡)'})
df = df.rename(columns={'접수년도': '접수연도'})
```

```python
new = pd.concat([df,df1])
```

```python
new.isnull().sum()
```

```python
new = new.dropna()
```

```python
new
```

```python
new = new[~new['건물명'].str.contains(r'\t\(\d+(-\d+)?\)\t')]
```

```python
new['본번'] = new['본번'].astype('int')
new['부번'] = new['부번'].astype('int')
new['층'] = new['층'].astype('int')
new['건축년도'] = new['건축년도'].astype('int')
```

```python
new['지번주소'] = '서울특별시 ' + new['자치구명']+' '+new['법정동명']+' '+new['본번'].astype('str')+'-'+new['부번'].astype('str')
```

```python
new['지번주소'].to_csv('지번주소.csv', index=False)
```
