```python
import pandas as pd
```

```python
df = pd.read_csv("매물 좌표.csv", encoding='cp949', index_col=0)
```

```python
df
```

```python
df
```

```python
mamul = pd.read_csv("mamul.csv")
```

```python
mamul = mamul[mamul['전월세구분'] == '월세']
```

```python
mamul.reset_index(drop=True, inplace=True)
```

```python
mamul = mamul.reset_index().rename(columns={'index': 'origin_index'})
```

```python
mamul
```

```python
# 이제 df1과 df2_reset을 'origin_index'를 기준으로 합칩니다.
merged_df = pd.merge(df, mamul, on='origin_index', how='inner')
```

```python
merged_df = merged_df.drop(axis=1, columns=['지번주소','origin_index'])
```

```python
merged_df
```

```python
merged_df['법정동명'].value_counts()
```

```python
# 'cordinates' 컬럼에서 괄호와 따옴표를 제거합니다.
coordinates_df = merged_df['cordinates'].str.replace("[()']", "", regex=True).str.split(',', expand=True)

# 분리된 데이터프레임의 컬럼 이름을 'x', 'y'로 지정합니다.
coordinates_df.columns = ['x', 'y']

# 원본 데이터프레임에 새로운 'x', 'y' 컬럼을 추가합니다.
merged_df['x'] = coordinates_df['x'].str.strip()  # 공백 제거
merged_df['y'] = coordinates_df['y'].str.strip()  # 공백 제거
```

```python
merged_df
```

```python
merged_df = merged_df.drop(axis=1, columns=['cordinates'])
merged_df
```

## 신사동(은평구)는 드랍, 신정동(마포구)는 드랍

```python
# 원래의 데이터프레임을 'merged_df'라고 가정합니다.
# '자치구명'이 '마포구'가 아니거나 '법정동명'이 '신정동'이 아닌 데이터만 필터링하여 선택합니다.
cleaned_df = merged_df[~((merged_df['자치구명'] == '마포구') & (merged_df['법정동명'] == '신정동'))]
cleaned_df = merged_df[~((merged_df['자치구명'] == '은평구') & (merged_df['법정동명'] == '신사동'))]

# 결과 확인
cleaned_df
```

```python
cleaned_df
```

## 본번 부번 접수연도 드랍해버리자

```python
cleaned_df.drop(axis=1, columns=['본번','부번','접수연도'], inplace=True)
```

```python
cleaned_df
```

## 법정동과 구를 pk로 바꿔주자

```python
gu = pd.read_csv("자치구.csv", encoding='cp949', sep=";")
gu.head(2)
```

```python
new = pd.merge(cleaned_df, gu, left_on='자치구명', right_on='gu_name', how='left')
```

```python
new
```

```python
dong = pd.read_csv("군집 결과.csv", index_col=0)
dong['dong_id'] = dong.index + 1
dong = dong[['법정동명','dong_id']]
dong
```

```python
new = pd.merge(new, dong, left_on='법정동명', right_on='법정동명', how='left')
new
```

## 자치구명, 법정동명, gu_name 드랍

```python
new = new.drop(axis=1, columns=['자치구명', '법정동명', 'gu_name'])
```

```python
new.rename(columns={'층': 'floor', '전월세구분' : 'buildingtype', '임대면적(㎡)' : 'area', '건물명' : 'name', '건축년도' : 'year'}, inplace=True)
```

```python
new.rename(columns={'buildingtype': 'pay_type', '건물용도' : 'building_type'}, inplace=True)
```

```python
new
```

```python
new.rename(columns={'보증금(만원)': 'deposit', '임대료(만원)' : 'monthly_pay'}, inplace=True)
```

```python
new.to_csv("final_dataset/mamul.csv", encoding='cp949', index=0)
```
