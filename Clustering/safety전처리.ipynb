## Reverse Geocoding module

```python
import requests, json
from tqdm.notebook import tqdm
from requests.adapters import HTTPAdapter
from urllib3.util.retry import Retry
```

```python
def get_address(coordinates):
    session = requests.Session()
    retry = Retry(connect=3, backoff_factor=0.5)
    adapter = HTTPAdapter(max_retries=retry)
    session.mount('http://', adapter)
    session.mount('https://', adapter)
    headers = {"Authorization": "KakaoAK ba25a018cf0921215c5f8c824762526e"}
    result = []
    
    for i, cord in tqdm(enumerate(coordinates)):
        lng = str(cord[1])
        lat = str(cord[0])
        url = "https://dapi.kakao.com/v2/local/geo/coord2regioncode.json?x="+lng+"&y="+lat
        api_json = session.get(url, headers=headers)
        result.append((i, json.loads(api_json.text)))

    return result
```

## 데이터 위도 및 경도 추출

```python
coordinates = list(zip(df1['WGS84위도'], df1['WGS84경도']))
```

## 모듈 실행

```python
result = get_address(coordinates)
result
```

## 법정동 데이터 추출

```python
preprocessing_result = []

for data in tqdm(result):
    try:
        d = [data[1]['documents'][0]['region_3depth_name'] + " " + data[1]['documents'][0]['region_4depth_name'],
             data[1]['documents'][0]['code']
            ]
        preprocessing_result.append(d)
    except KeyError:
        preprocessing_result.append(('Incorrect cordination', -1))
```
