```python
import pandas as pd
```

```python
df1 = pd.read_csv('dataset/편의시설 변경 버전.csv', index_col=0)
df1.head(3)
```

```python
df = pd.read_csv('dataset/safety/서울시 유흥주점영업 인허가 정보 (1).csv',encoding='cp949')
df.head(3)
```

```python
df.columns
```

```python
df = df[['도로명주소','상세영업상태명']]
```

```python
df['법정동'] = df['도로명주소'].str.extract('.*\(([^)]+)\)$')
df = df[['상세영업상태명','법정동']]
```

```python
df
```

```python
# '법정동' 컬럼에서 '동'으로 끝나는 부분만 추출
df['법정동'] = df['법정동'].str.extract('([가-힣]+동)')
```

```python
df = df[(df['상세영업상태명'] == '영업') & pd.notnull(df['법정동'])]
```

```python
df1
```

```python
df = df['법정동'].value_counts()
```

```python
merged_df = pd.merge(df1, df, left_on='법정동명', right_on='법정동', how='left')
```

```python
df.head(30)
```

```python
merged_df = merged_df.rename(columns={'count': '유흥주점'})
merged_df['유흥주점'].fillna(0, inplace=True)
merged_df
```

```python
merged_df.columns
```

```python
merged_df.to_csv('편의시설 변경 버전.csv', index=False)
```
