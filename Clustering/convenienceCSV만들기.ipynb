```python
import pandas as pd
```

```python
df = pd.read_csv("편의시설 변경 버전.csv", encoding='cp949')
```

```python
df['편의시설'] = df['마트_y'] + df['편의점']
```

```python
df = df[['법정동명','마트_y','편의점','편의시설']]
```

## 인덱스 1씩 미루기

```python
df['dong_id'] = df.index + 1
df
```

```python
df = df.drop(axis=1, columns=['법정동명'])
```

```python
df['마트_y'] = df['마트_y'].astype(int)
df['편의점'] = df['편의점'].astype(int)
df['편의시설'] = df['편의시설'].astype(int)
```

```python
df.to_csv("final_dataset/convenience.csv", encoding='cp949', index=0)
```
