```python
df = pd.read_csv("../군집 결과-class.csv", index_col=0).drop(axis=1, columns=['군집'])
```

```python
# 기존 데이터프레임 df를 가정합니다. 예를 들어:
# df = pd.DataFrame({
#     '법정동': ['청운동', '신교동', ...],
#     '치안': [2, 1, ...],
#     ...
# })

# 인프라 TypeID 매핑
type_id_mapping = {
    '치안': 2,
    '여가': 3,
    '일반음식점': 5,
    '보건': 4,
    '편의시설': 1,
    '대중교통': 8,
    '카페': 6,
    '술집': 7
}

# DongInfrastructureScore 데이터프레임 생성을 위한 준비
rows = []

# 주어진 데이터프레임의 각 행과 인프라 유형별 컬럼을 순회
for index, row in df.iterrows():
    dong_id = index + 1  # 동 ID는 인덱스 + 1로 가정
    for infra_type, type_id in type_id_mapping.items():
        score = row[infra_type]  # 해당 인프라 유형의 점수
        rows.append({'dong_id': dong_id, 'type_id': type_id, 'score': score})

# 새로운 DongInfrastructureScore 데이터프레임 생성
dong_infrastructure_score_df = pd.DataFrame(rows)

# 결과 확인
dong_infrastructure_score_df
```

```python
dong_infrastructure_score_df['score_id'] = dong_infrastructure_score_df.index + 1
dong_infrastructure_score_df.to_csv("infra_score.csv", encoding='cp949', index=0)
```

```python
df = pd.read_csv("../편의시설 변경 버전.csv", encoding='cp949')
df = df[['법정동명', '지하철역', '버스정류소',
       '영화상영관', '병원', '약국', '의원', '치안센터', '경찰서', '119안전센터',
       '유흥주점', '다방', '커피숍',
       '패스트푸드', '편의점', '헬스시설2', '술집', '음식점_y',
       '마트_y', 'CCTV수']]
df
```

```python
df = df.rename(columns={"음식점_y" : "음식점", "마트_y" : "마트", '헬스시설2' : "헬스시설"})
df
```

```python
# float64 타입인 컬럼을 int로 변경하는 코드
df = df.astype({
    '유흥주점': 'int64',
    '다방': 'int64',
    '커피숍': 'int64',
    '패스트푸드': 'int64',
    '편의점': 'int64',
    '헬스시설': 'int64',
    '술집': 'int64',
    '음식점': 'int64',
    '마트': 'int64',
    'CCTV수': 'int64'
})
```

```python
import pandas as pd

# 인프라 항목과 TypeID, ItemID 매핑
infra_items = {
    '편의점': (1, 15),
    '마트': (1, 14),
    '경찰서': (2, 1),
    '119안전센터': (2, 2),
    '치안센터': (2, 3),
    'CCTV수': (2, 4),
    '헬스시설': (3, 5),
    '영화상영관': (3, 6),
    '의원': (4, 13),
    '약국': (4, 12),
    '병원': (4, 11),
    '패스트푸드': (5, 10),
    '음식점': (5, 9),
    '커피숍': (6, 8),
    '다방': (6, 7),
    '술집': (7, 18),
    '유흥주점': (7, 19),
    '지하철역': (8, 16),
    '버스정류소': (8, 17)
}

# 기존 데이터프레임 df 예시
# 이 코드에서 df는 이미 생성되어 있다고 가정합니다.

# DongInfrastructureInfo 데이터프레임 생성을 위한 준비
rows = []

# 인덱스 + 1로 DongID 생성
df['DongID'] = df.index + 1

# 인프라 유형별로 각 항목의 개수를 합산하여 DongInfrastructureInfo에 추가
for infra_name, (type_id, item_id) in infra_items.items():
    for index, row in df.iterrows():
        count = row[infra_name]
        dong_id = row['DongID']
        # 개수가 0보다 큰 항목만 추가
        rows.append({'dong_id': dong_id, 'type_id': type_id, 'infra_id': item_id, 'count': count})

# 새로운 DongInfrastructureInfo 데이터프레임 생성
dong_infrastructure_info_df = pd.DataFrame(rows)

# 결과 확인
dong_infrastructure_info_df
```

```python
dong_infrastructure_info_df['count_id'] = dong_infrastructure_info_df.index + 1
dong_infrastructure_info_df.to_csv("infra_count.csv", encoding='cp949', index=0)
```
