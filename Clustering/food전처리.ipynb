```python
import pandas as pd
```

```python
df1 = pd.read_csv('편의시설 변경 버전.csv')
df1.head(3)
```

```python
df = pd.read_csv('dataset/food/서울시 휴게음식점 인허가 정보.csv', encoding='cp949', encoding_errors='ignore')
df
```

```python
df.columns
```

```python
df = df[['도로명주소','상세영업상태명', '사업장명','업태구분명']]
```

```python
df = df[df['상세영업상태명'] == '영업']
```

```python
df['업태구분명']
```

```python
df['법정동'] = df['도로명주소'].str.extract('.*\(([^)]+)\)$')
```

```python
df = df[pd.notnull(df['법정동'])]
```

```python
df
```

```python
# '법정동명' 컬럼에서 콤마로 구분된 첫 번째 항목만 추출
df['법정동명_추출'] = df['법정동'].apply(lambda x: x.split(',')[0])

# '법정동명_추출'이 '동', '로', '가'로 끝나는지 확인
df['법정동_종류'] = df['법정동명_추출'].apply(lambda x: '동' if x.endswith('동') else ('로' if x.endswith('로') else ('가' if x.endswith('가') else '기타')))
```

```python
df = df.drop(axis=1, columns='도로명주소')
```

```python
df[['업태구분명','법정동명_추출']]
```

```python
grouped_df = df.groupby(['법정동명_추출', '업태구분명']).size().reset_index(name='개수')
# grouped_df.sort_values(ascending=False, by='개수')
grouped_df
```

```python
df.groupby(['업태구분명']).size().reset_index(name='개수')
```

```python
grouped_df
```

```python
grouped_df[grouped_df['법정동명_추출'] == '영등포동1가']
```

```python
grouped_df['업태구분명'].unique()
```

```python
filtered_df = grouped_df[grouped_df['업태구분명'].isin(['편의점', '커피숍', '다방', '패스트푸드', '기타 휴게음식점', '일반조리판매'])]
filtered_df
```

```python
# '법정동'과 '업태구분명'으로 그룹화하고 '개수'의 합을 계산
grouped_df = filtered_df.groupby(['법정동명_추출', '업태구분명'])['개수'].sum().reset_index()

# 피벗 테이블 생성
pivot_table = grouped_df.pivot(index='법정동명_추출', columns='업태구분명', values='개수').fillna(0)
```

```python
reset_df = pivot_table.reset_index()
reset_df.head(2)
```

```python
reset_df
```

```python
merged_df = pd.merge(df1, reset_df, left_on='법정동명', right_on='법정동명_추출', how='left')
```

```python
merged_df
```

```python
merged_df.to_csv('편의시설 변경 버전.csv', index=False)
```

### 아래는 일반음식점

```python
import pandas as pd
```

```python
df = pd.read_csv('dataset/food/서울시 일반음식점 인허가 정보.csv', encoding='cp949', encoding_errors='ignore')
df
```

```python
df.columns
```

```python
df = df[['사업장명','상세영업상태명','도로명주소', '업태구분명']]
df = df[df['상세영업상태명'] == '영업']
df
```

```python
df['법정동'] = df['도로명주소'].str.extract('.*\(([^)]+)\)$')
```

```python
df = df[pd.notnull(df['법정동'])]
```

```python
df
```

```python
df['업태구분명'].unique()
```

```python
# '법정동명' 컬럼에서 콤마로 구분된 첫 번째 항목만 추출
df['법정동명_추출'] = df['법정동'].apply(lambda x: x.split(',')[0])
```

```python
df = df.dropna()
```

```python
df
```

```python
df['업태구분명'].value_counts()
```

```python
df = df[df['업태구분명'].isin(['정종/대포집/소주방', '감성주점', '라이브카페', '이동조리', '복어취급', '출장조리', '탕류(보신용)', '냉면집', '패밀리레스트랑', '뷔페식', '김밥(도시락)', '횟집', '식육(숯불구이)', '외국음식전문점(인도,태국등)', '통닭(치킨)', '중국식', '일식', '분식', '경양식', '호프/통닭', '기타', '한식'])]
```

```python
df['업태구분명'].value_counts().sort_values()
```

```python
# '업태구분명'을 기준으로 '술집', '음식점'으로 분류하는 함수 정의
def classify_shop(x):
    if x in ['라이브카페', '정종/대포집/소주방', '감성주점']:
        return '술집'
    else:
        return '음식점'
```

```python
df
```

```python
# 새로운 분류 컬럼 생성
df['분류'] = df['업태구분명'].apply(classify_shop)

# '법정동명_추출'과 '분류'를 기준으로 그룹화하고, '개수'의 합계를 구함
result_df = df.groupby(['법정동명_추출', '분류'])['사업장명'].count().reset_index(name='개수')
```

```python
pivot_df = result_df.pivot_table(index='법정동명_추출', columns='분류', values='개수', fill_value=0).reset_index()
```

```python
pivot_df
```

```python
merged_df = pd.merge(df1, pivot_df, left_on='법정동명', right_on='법정동명_추출', how='left')
merged_df
```

```python
merged_df.columns
```

```python
merged_df[['음식점수', '음식점_x','음식점_y']]
```

```python
merged_df = merged_df.fillna(0)
```

```python
merged_df
```

```python
merged_df.to_csv('편의시설 변경 버전.csv', index=False)
```
