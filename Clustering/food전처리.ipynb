```python
import pandas as pd
```

```python
df1 = pd.read_csv('편의시설 변경 버전.csv')
df1.head(3)
```

```python
df = pd.read_csv('dataset/food/서울시 휴게음식점 인허가 정보.csv', encoding='cp949', encoding_errors='ignore')
df
```

```python
df.columns
```

```python
df = df[['도로명주소','상세영업상태명', '업태구분명']]
```

```python
df = df[df['상세영업상태명'] == '영업']
```

```python
df['업태구분명']
```

```python
df['도로명주소']
```

```python
df['법정동'] = df['도로명주소'].str.extract('.*\(([^)]+)\)$')
```

```python
df = df[pd.notnull(df['법정동'])]
```

```python
df['법정동']
```

```python
df
```

```python
df = df.drop(axis=1, columns='도로명주소')
```

```python
df
```

```python
# '법정동' 컬럼에서 '동'으로 끝나는 부분만 추출
df['법정동'] = df['도로명주소'].str.extract('([가-힣]+동)')
```

```python
df[['업태구분명','법정동']]
```

```python
grouped_df = df.groupby(['법정동', '업태구분명']).size().reset_index(name='개수')
# grouped_df.sort_values(ascending=False, by='개수')
grouped_df
```

```python
grouped_df['법정동'] = grouped_df['법정동'].str.split(',').str[0]
```

```python
grouped_df
```

```python
grouped_df[grouped_df['법정동'] == '영등포동1가']
```

```python
filtered_df = grouped_df[grouped_df['업태구분명'].isin(['편의점', '커피숍', '다방', '패스트푸드'])]
filtered_df
```

```python
# '법정동'과 '업태구분명'으로 그룹화하고 '개수'의 합을 계산
grouped_df = filtered_df.groupby(['법정동', '업태구분명'])['개수'].sum().reset_index()

# 피벗 테이블 생성
pivot_table = grouped_df.pivot(index='법정동', columns='업태구분명', values='개수').fillna(0)
```

```python
reset_df = pivot_table.reset_index()
reset_df.head(2)
```

```python
merged_df = pd.merge(df1, reset_df, left_on='법정동명', right_on='법정동', how='left')
```

```python
merged_df = merged_df.drop(axis=1, columns='법정동')
merged_df
```

```python
merged_df.columns
```

```python
merged_df.to_csv('편의시설 변경 버전.csv', index=False)
```

### 아래는 일반음식점

```python
import pandas as pd
```

```python
df = pd.read_csv('dataset/food/서울시 일반음식점 인허가 정보.csv', encoding='cp949', encoding_errors='ignore')
df
```

```python
df.columns
```

```python
df = df[['사업장명','상세영업상태명','도로명주소', '업태구분명']]
df = df[df['상세영업상태명'] == '영업']
df
```

```python
df['법정동'] = df['도로명주소'].str.extract('.*\(([^)]+)\)$')
```

```python
df = df[pd.notnull(df['법정동'])]
```

```python
df['법정동'] = df['법정동'].str.split(',').str[0]
```

```python
df['업태구분명'].unique()
```

```python
df = df.dropna()
```

```python
df = df[df['업태구분명'].isin(['한식','일식','경양식','중국식','외국음식전문점(인도,태국등)','식육(숯불구이)', '분식', '김밥(도시락)','정종/대포집/소주방','횟집','패밀리레스토랑','뷔페식','냉면집','통닭(치킨)'])]
```

```python
grouped_df = df.groupby(['법정동', '업태구분명']).size().reset_index(name='개수')
```

```python
grouped_df = grouped_df.groupby(['법정동'])['개수'].sum().reset_index()

# 피벗 테이블 생성
# pivot_table = grouped_df.pivot(index='법정동', columns='업태구분명', values='개수').fillna(0)
```

```python
grouped_df
```

```python
merged_df = pd.merge(df1, grouped_df, left_on='법정동명', right_on='법정동', how='left')
merged_df
```

```python
merged_df = merged_df.fillna(0)
```

```python
merged_df = merged_df.drop(axis=1, columns=['법정동'])
```

```python
merged_df = merged_df.rename(columns={'개수': '음식점'})
```

```python
merged_df
```

```python
merged_df.to_csv('편의시설 변경 버전.csv', index=False)
```
