```python
import pandas as pd
```

```python
df1 = pd.read_csv('dataset/서울교통공사 지하철역 주소 및 전화번호 정보.csv', encoding='cp949')
```

```python
df1 = df1[['호선','역명','도로명주소']]
df1['도로명주소1'] = df1['도로명주소']
```
