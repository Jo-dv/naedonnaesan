```python
import pandas as pd

# 주어진 인덱스 데이터
index_data = ['1111010100', '296', '서울특별시', '종로구', '청운동', 'Unnamed: 5', '37.587111,126.969069']

# 새로운 컬럼 이름
new_columns = ['법정동코드', '주소수', '시도', '시군구', '읍면동', '리', '중심좌표']

# 인덱스 데이터를 데이터프레임의 행으로 변환
# 여기서 index_data는 리스트 형태로, 각 요소가 새 데이터프레임의 컬럼 값이 됩니다.
df2 = pd.DataFrame([index_data], columns=new_columns)

print(df2)
```

```python
# 새로운 컬럼 이름 설정


# 현재 데이터프레임의 첫 번째 행을 새로운 컬럼 이름으로 사용합니다.
df.columns = new_columns

new = pd.concat([df2, df], ignore_index=True)
# # 데이터프레임의 인덱스를 재설정합니다.
# df.reset_index(drop=True, inplace=True)

# print(df)
new
```

```python
new = new[new['시도'] == '서울특별시']
new
```

```python
# '중심좌표' 컬럼을 ','로 분리하고, 결과 리스트의 첫 번째와 두 번째 요소를 각각 새로운 컬럼에 할당합니다.
new[['위도', '경도']] = new['중심좌표'].str.split(',', expand=True)
```

```python
new = new.drop(axis=1, columns=['중심좌표'])
```

```python
new = new.drop(axis=1, columns=['리','법정동코드', '주소수', '시도'])
```

```python
df = pd.read_csv("데이터 넣기용 동.csv")
df
```

```python
merged = pd.merge(df, new, left_on='dong_name', right_on='읍면동', how='left' )
```

```python
merged.isnull().sum()
```

```python
merged = merged.drop(axis=1, columns=['읍면동'])
```

```python
gu = pd.read_csv("자치구.csv", sep=';', encoding='cp949')
```

```python
merged = merged.rename(columns={'시군구' : 'gu_id', '위도' : 'dongx', '경도' : 'dongy'})
```

```python
# 두 번째 데이터프레임을 이용해 구 이름을 ID로 매핑하는 딕셔너리 생성
gu_name_to_id = pd.Series(gu.gu_id.values,index=gu.gu_name).to_dict()

# 첫 번째 데이터프레임의 gu_id 컬럼을 구 이름에서 ID로 변경
merged['gu_id'] = merged['gu_id'].map(gu_name_to_id)
```
