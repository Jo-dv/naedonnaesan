```python
import pandas as pd

# 주어진 인덱스 데이터
index_data = ['1111010100', '296', '서울특별시', '종로구', '청운동', 'Unnamed: 5', '37.587111,126.969069']

# 새로운 컬럼 이름
new_columns = ['법정동코드', '주소수', '시도', '시군구', '읍면동', '리', '중심좌표']

# 인덱스 데이터를 데이터프레임의 행으로 변환
# 여기서 index_data는 리스트 형태로, 각 요소가 새 데이터프레임의 컬럼 값이 됩니다.
df2 = pd.DataFrame([index_data], columns=new_columns)

print(df2)
```
